```python
import pandas as pd

nssp = pd.read_csv('nssp_ed_demand_weekly_michigan_curated.csv')
nhsn = pd.read_csv('nhsn_capacity_weekly_mi_curated.csv')

nssp['week_end'] = pd.to_datetime(nssp['week_end'])
nhsn['week_end'] = pd.to_datetime(nhsn['week_end'])
```

```python
nssp['week_end'].min(), nssp['week_end'].max()
```

```python
nhsn['week_end'].min(), nhsn['week_end'].max()
```

```python
analysis_weekly = (
    nssp.merge(
        nhsn,
        on='week_end',
        how='inner'
    )
    .sort_values('week_end')
)
```

```python
import os
os.makedirs('data_analysis', exist_ok=True)

analysis_weekly.to_csv(
    'data_analysis/analysis_ed_demand_vs_capacity_weekly_mi.csv',
    index=False
)
```
