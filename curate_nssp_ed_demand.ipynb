```python
import pandas as pd
import numpy as np
```

```python
nhsn = pd.read_csv("NSSP_Emergency_Department_Visit_Trajectories_by_State_and_Sub_State_Regions-_COVID-19__Flu__RSV__Combined__.csv")
```

```python
nhsn.head()
```

```python
mi = nhsn[nhsn['geography'] == 'Michigan'].copy()
mi.head()
```

```python
mi['week_end'] = pd.to_datetime(mi['week_end'])

mi_state = (
    mi.groupby('week_end', as_index=False)
      .agg({
          'percent_visits_smoothed_combined': 'mean',
          'percent_visits_smoothed_influenza': 'mean',
          'percent_visits_smoothed_covid': 'mean'
      })
      .sort_values('week_end')
)

mi_state.head()
```

```python
mi_state.describe()
```

```python
mi_state.shape
```

```python
mi_state_2023 = mi_state[mi_state['week_end'] >= '2023-01-01'].copy()
mi_state_2023.shape
```

```python
mi_state_2023.head()
```

```python
mi_state_2023 = mi_state_2023[
    mi_state_2023['week_end'] <= '2024-10-05'
].copy()

mi_state_2023.shape
```

```python
mi_state_2023.to_csv(
    'data_curated/nssp_ed_demand_weekly_mi_curated.csv',
    index=False
)
```
