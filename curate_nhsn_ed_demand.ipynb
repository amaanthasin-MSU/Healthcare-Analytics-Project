```python
import pandas as pd

nhsn = pd.read_csv('nhsn_weekly_hospital_capacity_michigan_raw.csv')

nhsn.head()
```

```python
nhsn.columns.tolist()
```

```python
nhsn_keep = [
    'Week Ending Date',
    'Geographic aggregation',
    'Percent Hospital Reporting Days',
    'Weekly Average Inpatient Beds',
    'Weekly Average ICU Beds',
    'Weekly Average Inpatient Beds Occupied',
    'Weekly Average ICU Beds Occupied',
    'Weekly Average Percent Inpatient Bed Occupancy',
    'Weekly Average Percent ICU Bed Occupancy',
    'Weekly Average Prevalent Influenza Hospitalizations',
    'Weekly Average Prevalent COVID-19 Hospitalizations',
    'Weekly Average Hospitalized COVID-19 ICU Patients'
]

nhsn_cur = nhsn[nhsn_keep].copy()
nhsn_cur.head()
```

```python
nhsn_cur.columns.tolist()
```

```python
nhsn_cur['Week Ending Date'] = pd.to_datetime(nhsn_cur['Week Ending Date'])
nhsn_cur['Percent Hospital Reporting Days'] = (
    nhsn_cur['Percent Hospital Reporting Days']
        .astype(str)
        .str.replace('%','')
        .astype(float)
)
```

```python
nhsn_filt = nhsn_cur[
    (nhsn_cur['Geographic aggregation'] == 'MI') &
    (nhsn_cur['Percent Hospital Reporting Days'] >= 90) &
    (nhsn_cur['Week Ending Date'] >= '2023-01-01') &
    (nhsn_cur['Week Ending Date'] <= '2024-10-05')
].copy()

nhsn_filt.head()
```

```python
nhsn_filt.shape
nhsn_filt['Week Ending Date'].min(), nhsn_filt['Week Ending Date'].max()
nhsn_filt.isna().sum()
```

```python
nhsn_final = nhsn_filt.rename(columns={
    'Week Ending Date': 'week_end',
    'Geographic aggregation': 'geo',
    'Percent Hospital Reporting Days': 'pct_reporting_days',

    'Weekly Average Inpatient Beds': 'avg_inpatient_beds',
    'Weekly Average ICU Beds': 'avg_icu_beds',
    'Weekly Average Inpatient Beds Occupied': 'avg_inpatient_beds_used',
    'Weekly Average ICU Beds Occupied': 'avg_icu_beds_used',

    'Weekly Average Percent Inpatient Bed Occupancy': 'pct_inpatient_beds_occupied',
    'Weekly Average Percent ICU Bed Occupancy': 'pct_icu_beds_occupied',

    'Weekly Average Prevalent Influenza Hospitalizations': 'avg_flu_hospitalizations',
    'Weekly Average Prevalent COVID-19 Hospitalizations': 'avg_covid_hospitalizations',
    'Weekly Average Hospitalized COVID-19 ICU Patients': 'avg_covid_icu_patients'
}).copy()
```

```python
for c in ['pct_inpatient_beds_occupied', 'pct_icu_beds_occupied']:
    nhsn_final[c] = nhsn_final[c].astype(str).str.replace('%','').astype(float)

nhsn_final.head()
```

```python
nhsn_final.to_csv(
    'data_curated/nhsn_capacity_weekly_mi_curated.csv',
    index=False
)
```
